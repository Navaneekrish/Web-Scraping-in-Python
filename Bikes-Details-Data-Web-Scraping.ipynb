```python
from bs4 import BeautifulSoup
import requests,openpyxl
```

```python
excel_file = openpyxl.Workbook()
print(excel_file.sheetnames)
sheets =excel_file.active
sheets.title = "Bike list"
print(excel_file.sheetnames)
sheets.append(['Name','Price','Rating','No of Reviews','Body_type','No_of_seats','Mileage KMPL or KMFC'])
```

```python
with open('bikes_data.html', 'r', encoding="utf8") as f:
    html_content = f.read()

try:
    soup = BeautifulSoup(html_content, 'html.parser')

    bikes = soup.find('div', class_="row ajax_scroll_search mt10")

    # Check if the 'div' element is found
    if bikes:
        bike_list = bikes.find_all('div', class_='details')  #common class for bike details

        for bike in bike_list:
            name_elements = bike.find_all('h3')
            for name_element in name_elements:
                name = name_element.text

            price = bike.find('h4').get_text(strip=True).split('Onwards')[0]
            
            rating = float(bike.find('span').text)
            
            no_of_review = int(bike.find('div',class_="reviews").get_text(strip=True).split(" ")[0])
            
            Body_type = bike.find('div',class_='info').span.get_text(strip=True).split(',')[0]
            
            No_of_seats = bike.find('div',class_='info').span.get_text(strip=True).split(',')[1]
            seat = int(No_of_seats.split(' ')[-2])
            
            mileage = bike.find('div',class_='info').span.get_text(strip=True).split(',')[2]
            mil = int(mileage.split(' ')[-2])
            
            
            print(name,price,rating,no_of_review,Body_type,seat,mil)
            sheets.append([name,price,rating,no_of_review,Body_type,seat,mil])
    
    else:
        print("Could not find the 'div' element with class 'row ajax_scroll_search mt10'.")

except Exception as e:
    print(e)

excel_file.save('list of Bike Details.xlsx')
```
